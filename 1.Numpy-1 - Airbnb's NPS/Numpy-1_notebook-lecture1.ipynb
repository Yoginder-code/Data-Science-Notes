```python
birds = ['spoonbills',  'plovers',  'plovers',  'plovers', 
         'plovers',  'Cranes',  'plovers',  'plovers',  'Cranes',  'spoonbills']
age = [5.5, 6.0, 3.5, 1.5, 3.0, 4.0, 3.5, 2.0, 5.5, 6.0]
```

```python
a = []
for i in range(len(birds)):
    if birds[i] == 'Cranes':
        a.append(i)
```

```python
a
```

```python
c=[]
for i in a:
    for j in range(len(age)):
        if i == j:
            c.append(age[j])
```
