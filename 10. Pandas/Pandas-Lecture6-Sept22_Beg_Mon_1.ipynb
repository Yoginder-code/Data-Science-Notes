```python
def temp_mean(x):
    x["Temp_avg"]=np.round(x["Temperature"].mean(),2)
    return x
```

```python
df_tidy=df_tidy.groupby("Drug_Name").apply(temp_mean)
df_tidy
```

```python
df_tidy["Temperature"].fillna(df_tidy["Temp_avg"],inplace=True)
```

```python
df_tidy
```

```python

def pres_mean(x):
    x["Pres_avg"]=np.round(x["Pressure"].mean(),2)
    return x
```

```python
df_tidy=df_tidy.groupby("Drug_Name").apply(pres_mean)
```

```python
df_tidy
```

```python
df_tidy["Pressure"].fillna(df_tidy["Pres_avg"],inplace=True)
df_tidy
```
